# 방전구간 데이터 분석 (2022년 4월 ~ 8월)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [20, 4] # setting figure size

# 통합하고자 하는 모든 데이터의 연도/월 (각 파일명 참고) 나열
year_month_list = ['2204', '2205', '2206','2207', '2208']
# year_month_list = ['2021', '2022']
```

### 데이터 개요

```python
df_concat_list = []
for year_month in year_month_list:
    df_concat_list.append(pd.read_csv('./data/' + year_month + '_discharge_merged_statistics.csv'))
    
df = pd.concat(df_concat_list, axis=0, ignore_index=True)
df.sort_values(by=['date_start'], ascending=True, inplace=True)
df.reset_index(drop=True, inplace=True)

groups = df.groupby('car_type')

print("차종별 급속충전 구간 수")
for name, group in groups:
    print("%6s : %5d" % (name, len(group)))

df_niro = df[df['car_type'].str.contains('(?i)niro')]
df_ioniq5 = df[df['car_type'].str.contains('(?i)ioniq5')]
df_ev6 = df[df['car_type'].str.contains('(?i)ev6')]
df_kona = df[df['car_type'].str.contains('(?i)kona')]
df_g80 = df[df['car_type'].str.contains('(?i)g80')]

car_type_list = ['', '_niro', '_ioniq5', '_ev6', '_kona', '_g80']

for car_type in car_type_list:
    if len(globals()[f'df'+car_type]) == 0:
        car_type_list.remove(car_type)
        continue
    globals()[f'df'+car_type].reset_index(drop=True, inplace=True)


print("\n데이터 칼럼 수 : %d" % len(df.columns))
print(list(df.columns))
    
df
```

#### - 월별 방전 속도, 전류 평균 그래프

```python
for car_type in car_type_list:
    globals()[f'df04'+car_type] = globals()[f'df'+car_type].query("date_start > '2022-03-32' and date_start < '2022-05'") #4월
    globals()[f'df05'+car_type] = globals()[f'df'+car_type].query("date_start > '2022-04-31' and date_start < '2022-06'") #5월
    globals()[f'df06'+car_type] = globals()[f'df'+car_type].query("date_start > '2022-05-32' and date_start < '2022-07'") #6월
```

```python
plt.rcParams['figure.figsize'] = [20, 4] # setting figure size

for car_type in car_type_list:
        fig, ax = plt.subplots()
        ax.plot(globals()[f'df04'+car_type].index,  # x
        globals()[f'df04'+car_type].discharge_speed,  # y 
                marker='.', 
                markersize=5,
                linestyle='',
                label=4)
        ax.plot(globals()[f'df05'+car_type].index,  # x
        globals()[f'df05'+car_type].discharge_speed,  # y 
                marker='.', 
                markersize=5,
                linestyle='',
                label=5)
        ax.plot(globals()[f'df06'+car_type].index,  # x
        globals()[f'df06'+car_type].discharge_speed,  # y 
                marker='.', 
                markersize=5,
                linestyle='',
                label=6)
        ax.legend(fontsize=12, loc='upper right') # legend position
        plt.title('discharging speed by Month' + car_type, fontsize=20)
        plt.xlabel('discharging section index', fontsize=14)
        plt.ylabel('discharging speed', fontsize=14)
        plt.ylim(0, 50)
        plt.grid()
        plt.show()

        sns.boxplot(x = globals()[f'df'+car_type]['date_start'].str[5:7], y = globals()[f'df'+car_type]['discharge_speed'], order = ['04', '05', '06']).set(xlabel='month')
        #print("%7s   %.2f   %3d   %3d   %3d" % (name, group['total_collection_time_hour'].mean(), group['total_collection_time_hour'].max(),group['total_collection_time_hour'].min(),len(group)))
        plt.title('discharging speed by Month_boxplot' + car_type, fontsize=20)
        plt.ylim(0, 50)
        plt.grid()
        plt.show()
```

```python
plt.rcParams['figure.figsize'] = [20, 4] # setting figure size

for car_type in car_type_list:
        fig, ax = plt.subplots()
        ax.plot(globals()[f'df04'+car_type].index,  # x
        globals()[f'df04'+car_type].current_avg,  # y 
                marker='.', 
                markersize=5,
                linestyle='',
                label=4)
        ax.plot(globals()[f'df05'+car_type].index,  # x
        globals()[f'df05'+car_type].current_avg,  # y 
                marker='.', 
                markersize=5,
                linestyle='',
                label=5)
        ax.plot(globals()[f'df06'+car_type].index,  # x
        globals()[f'df06'+car_type].current_avg,  # y 
                marker='.', 
                markersize=5,
                linestyle='',
                label=6)
        ax.legend(fontsize=12, loc='upper right') # legend position
        plt.title('current_avg by Month' + car_type, fontsize=20)
        plt.xlabel('discharging section index', fontsize=14)
        plt.ylabel('current_avg', fontsize=14)
        plt.ylim(-20, 80)
        plt.grid()
        plt.show()

        sns.boxplot(x = globals()[f'df'+car_type]['date_start'].str[5:7], y = globals()[f'df'+car_type]['current_avg'], order = ['04', '05', '06']).set(xlabel='month')
        #print("%7s   %.2f   %3d   %3d   %3d" % (name, group['total_collection_time_hour'].mean(), group['total_collection_time_hour'].max(),group['total_collection_time_hour'].min(),len(group)))
        plt.title('current_avg by Month_boxplot' + car_type, fontsize=20)
        plt.ylim(-20, 50)
        plt.grid()
        plt.show()
```

#### - 각 방전구간별 전류 평균 그래프

```python
fig, ax = plt.subplots()
for name, group in groups:
    ax.plot(group.index,  # x
             group.current_avg,  # y 
            marker='.', 
            markersize=3,
            linestyle='',
            label=name)
ax.legend(fontsize=12, loc='upper right') # legend position
plt.title('Current AVG per discharging section', fontsize=20)
plt.xlabel('discharging section index', fontsize=14)
plt.ylabel('current average', fontsize=14)
plt.show()
```

## 외기온도 범위에 따른 방전 속도 분석

```python
df_temp_1 = df.query("extern_temp_avg > -10 and extern_temp_avg <= 0")
df_temp_2 = df.query("extern_temp_avg > 0 and extern_temp_avg <= 10")
df_temp_3 = df.query("extern_temp_avg > 10 and extern_temp_avg <= 20")
df_temp_4 = df.query("extern_temp_avg > 20 and extern_temp_avg <= 30")
print(df_temp_1, df_temp_2, df_temp_3, df_temp_4)
```
